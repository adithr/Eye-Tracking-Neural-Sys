### Speed vs Noise

The following cells plot the speed of gaze against the perceptual noise of the samples in each game for a particular participant. The velocity is averaged over 16 trials for a particular game and so is the perceptual noise.

The speed is obtained by getting the velocities in directions x and y and then taking a geometric mean of it. The fixation positions x,y and the fixation time are obtained from the Efix data (https://github.com/esdalmaijer/PyGazeAnalyser/blob/master/pygazeanalyser/gazeplotter.py) which are then used to calculate the displacements and ultimately the velocity and speed.

The perceptual noise is available from the dataset_0423.csv file (https://github.com/adithr/Eye-Tracking-Neural-Sys). 



```python
# list participant ids
subject_ids = np.arange(30,37,1)
partic_ids = np.array(['G0'+str(id) for id in subject_ids]) #for filename


dataset_fname = 'dataset_0423.csv' #for perceptual noise


#define figures for plotting
row_plot = int(np.ceil(subject_ids.size/2))
col_plot = subject_ids.size - row_plot
fig,axes = plt.subplots(row_plot,col_plot,figsize = (30,30))



for i, partic_id in enumerate(partic_ids):
    try:
        ax = axes.flatten()[i] #flatten axes and return
        plot_speed_noise(partic_id, dataset_fname = 'dataset_0423.csv',mode='mean',fig=fig,
                                  ax = ax)
        
        #axes[i] =  ax

    except TypeError as e:
        print(e, 'participant: ',partic_id)
        continue
```

#### Comments

- At high perceptual noise, the speed is 'expected' to be high as well, but what we observe here is a slight increase and then a decrease for most participants
- We observe what looks like to be a local maximum of speed. This could point to another hypothesis that the participants are willing to explore for up to a certain noise but 'give up' beyond that certain point.    
- We think magnitude of acceleration (rate of change of speed) would be another metric to explore to see some patterns
